# Bank Simulation GUI
The Graphical User interface will function in the following way:
1) It will have a welcome page with a username and a password entry field and a log in button.

2) After entering the username and password, the user will be directed to a main menu page with options to create deposit money, withdraw money, check balance, and exit.

3) On the welcome page there will also be a sign up button. When the user clicks the sign up button, it will direct them to a new window that will require them to add their username, password and a confirmed password. The information collected here will be stored in an accounts.txt file that has the name, password and balance for each user.

```python
# importing the needed libraries
from tkinter import *
from tkinter import messagebox

# Function to open a new window for the sign-up form
"""
This function opens a child window for of 'root'.
It has a sign up form where the user is suppose to enter the username, password and then confirm the password.
This function runs when the user clicks sign up on the welcome page.
"""
def open_signup_window():
    # Creating a new window
    signup_window = Toplevel(root)
    # adding the title and the size for the sign up window
    signup_window.title("Sign Up")
    signup_window.geometry('400x300')
    
    # Creating labels and entry fields for the sign-up form
    # for the username
    username_label = Label(signup_window, text="Username:")
    username_label.grid(row=0, column=0, padx=10, pady=10, sticky=E)
    username_entry = Entry(signup_window, width=30)
    username_entry.grid(row=0, column=1, padx=10, pady=10)
    
    # for the password
    password_label = Label(signup_window, text="Password:")
    password_label.grid(row=1, column=0, padx=10, pady=10, sticky=E)
    password_entry = Entry(signup_window, width=30, show='*')
    password_entry.grid(row=1, column=1, padx=10, pady=10)

    # for the confirmed password
    confirm_password_label = Label(signup_window, text="Confirm Password:")
    confirm_password_label.grid(row=2, column=0, padx=10, pady=10, sticky=E)
    confirm_password_entry = Entry(signup_window, width=30, show='*')
    confirm_password_entry.grid(row=2, column=1, padx=10, pady=10)

    # Function to handle the sign-up process
    """
    This function gets the information entered in the username, password and confirm password field.
    It first checks if either field is empty and raises an error if so.
    Then it checks if the password and confirm password are the same, if not it raises an error.
    Finally, it checks if the username already exists and if it doesn't, it creates a new account.
    After the account is created, the window closes. You have to sign in again to perform any operation.
    This function is triggered when the submit button in the child window is pressed.
    """
    def sign_up():
        balance = 0.0
        # getting the information from the entry field
        username = username_entry.get().strip()
        password = password_entry.get().strip()
        confirm_password = confirm_password_entry.get().strip()

        # checking if either fields is empty
        if not username or not password or not confirm_password:
            messagebox.showerror('Error', 'Please fill in all fields.')
            return

        # checking if the passwords are the same
        if password != confirm_password:
            messagebox.showerror('Error', 'Passwords do not match.')
            return

        # Checking if username already exists
        with open("accounts.txt", 'r') as account_file:
            for line in account_file:
                if line.startswith(username + ':'):
                    messagebox.showerror('Error', 'Username already taken.')
                    return

        # If the username does not exist, create the account
        with open("accounts.txt", 'a') as account_file:
            account_file.write(f"{username}:{password}:{balance}\n")
            messagebox.showinfo('Success', 'Account created successfully!')
        
        # closing the window after the account is created
        signup_window.destroy() 

    # Creating a submit button for the sign up form
    submit_button = Button(signup_window, text="Submit", command=sign_up)
    submit_button.grid(row=3, columnspan=2, pady=10)

# Function to log in
"""
This function gets the info from the entries for the password and username.
It checks if the username exists and if the password is correct.
If both are correct, it opens another window that displays your current balance, and another field where the user enters the amount of money they
would want to withdraw or deposit.
Once the amount is entered, the user can either click deposit or withdraw. 
The function has another function that is triggered when the user clicks deposit or withdraw.
The log in function is triggered when the user presses the log in button.
"""
def log_in():
    global balance
    login_username = login_username_entry.get().strip()
    login_password = login_password_entry.get().strip()
    
    # Checking if the username and password are correct
    with open("accounts.txt", 'r') as account_file:
        for line in account_file:
            user, passw, bal = line.strip().split(':')
            if user == login_username and passw == login_password:
                balance = float(bal)
                messagebox.showinfo('Success', 'Login successful!')
                
                # Creating a new window for the logged in user
                logged_in_window = Toplevel(root)
                logged_in_window.title('Logged In')
                logged_in_window.geometry('400x400')
                logged_in_label = Label(logged_in_window, text=f'Welcome, {login_username}')
                logged_in_label.grid(pady=10, padx=10)

                # creating a current balance label
                current_balance_label = Label(logged_in_window, text=f'Current Balance: ${balance}')
                current_balance_label.grid(row=1, column=0, padx=10, pady=10)

                # creating an entry field for the amount of money the user wants to withdraw or deposit
                amount_label = Label(logged_in_window, text='Enter amount you wish to withdraw/deposit:$')
                amount_label.grid(row=2, column=0, padx=10, pady=10)
                amount_entry = Entry(logged_in_window, width=30)
                amount_entry.grid(row=2, column=1, padx=10, pady=10)

                # Define deposit function
                """
                This function is triggered when the user presses the deposit button.
                The function gets the amount entered and adds it to the existing balance.
                It finally updates the account.txt file with the new balances.
                """
                def deposit():
                    amount = float(amount_entry.get().strip())
                    new_balance = balance + amount
                    update_balance(login_username, login_password, new_balance)
                    messagebox.showinfo('Success', f'Deposited ${amount}')
                    logged_in_window.destroy()

                # Define withdraw function
                """
                This function is triggered when the user presses the withdraw button.
                The function gets the amount entered and subtracts it from the existing balance.
                If the amount entered in more than the current balance, an error is raise.
                It finally updates the account.txt file with the new balances.
                """
                def withdraw():
                    amount = float(amount_entry.get().strip())
                    if balance < amount:
                        messagebox.showerror('Error', 'Insufficient funds')
                    else:
                        new_balance = balance - amount
                        update_balance(login_username, login_password, new_balance)
                        messagebox.showinfo('Success', f'Withdrew ${amount}')
                        logged_in_window.destroy()

                # creating two buttons, for depositing and withdrawing
                deposit_button = Button(logged_in_window, fg='green', text='Deposit', command=deposit)
                deposit_button.grid(row=3, column=0, padx=10, pady=10)

                withdraw_button = Button(logged_in_window, fg='blue', text='Withdraw', command=withdraw)
                withdraw_button.grid(row=3, column=1, padx=10, pady=10)

                # creating an exit button - it destroys the log in window
                exit_button = Button(logged_in_window, fg='red', text='Exit', command=logged_in_window.destroy)
                exit_button.grid(row=4, columnspan=3, padx=10, pady=10)
                return
        
        messagebox.showerror('Error', 'Incorrect Username or Password!')

# Function to update the balance in the accounts file
def update_balance(username, password, new_balance):
    lines = []
    with open("accounts.txt", 'r') as account_file:
        lines = account_file.readlines()
    
    with open("accounts.txt", 'w') as account_file:
        for line in lines:
            user, passw, bal = line.strip().split(':')
            if user == username and passw == password:
                account_file.write(f"{username}:{password}:{new_balance}\n")
            else:
                account_file.write(line)

# creating a root window
root = Tk()
root.title('Bank Simulation: Trial 1')
root.geometry('400x400')

# Creating a welcome label
welcome_label = Label(root, text="Welcome to the Virtual Bank Of Kenya.")
welcome_label.grid(pady=20)

# Creating a sign up button for the root window
sign_up_button = Button(root, text="Sign Up", fg='blue', command=open_signup_window)
sign_up_button.grid(row=4, columnspan=2, padx=10, pady=10)

# Creating a username and password entry and label for login
login_username_label = Label(root, text="Username:")
login_username_label.grid(row=1, column=0, padx=10, pady=10)
login_username_entry = Entry(root, width=30)
login_username_entry.grid(row=1, column=1, padx=10, pady=10)

login_password_label = Label(root, text='Password:')
login_password_label.grid(row=2, column=0, padx=10, pady=10)
login_password_entry = Entry(root, width=30, show='*')
login_password_entry.grid(row=2, column=1, padx=10, pady=10)

# adding a sign in button in the root window
sign_in_button = Button(root, text='Log in', fg='green', command=log_in)
sign_in_button.grid(row=3, columnspan=2, pady=10)

# Run the main loop
root.mainloop()
```
